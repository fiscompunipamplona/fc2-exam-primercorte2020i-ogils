 $\textbf{Calculo de la covarianza $cov(x,y)$=$\overline{xy}$-$\overline{x}$$\overline{y}$,   y la correlacion $\rho=\frac{cov(x,y)}{\sigma_{x}\sigma_{y}}$ , aplicado a dos columnas de los datos.}$

```
w=datos[:,9] # masa de la estrella; en unidades de masa solar
ww=datos[:,10]  # radio de la estretlla; en unidades de radio solar
plot(w,'.r')
plot(ww,'.g')
grid()
show()
```

```
xmedia_ymedia=j.medial()*k.medial()
productoxy=w*ww
totalxy=sum(productoxy)
Nxy=len(productoxy)
xymedia=totalxy/Nxy
covarianzaxy=xymedia-xmedia_ymedia
correlacion=covarianzaxy/(j.destandar()*k.destandar())
print("\n covarianza...:",format(covarianzaxy,'.4f'),"\ncorrelacion...:" ,format(correlacion,'.4f'))
```

```
x=datos[:,0] # período orbital
y=datos[:,5] # radio del planeta; en unidades del radio de Júpiter
plot(x,'.b')
plot(y,'.y')
grid()
show()
```

```
xmedia_ymedia=a.medial()*e.medial()
productoxy=x*y
totalxy=sum(productoxy)
Nxy=len(productoxy)
xymedia=totalxy/Nxy
covarianzaxy=xymedia-xmedia_ymedia
correlacion=covarianzaxy/(a.destandar()*e.destandar())
print("\n covarianza...:",format(covarianzaxy,'.4f'),"\ncorrelacion...:" ,format(correlacion,'.4f'))
```

el resultado de la covarianza entre estas dos variables (período orbital y radio del planeta; en unidades del radio de Júpiter) es negativa entonces no existe una correlacion directa enre estas dos, es decir existe es una correlacion inversa, como podemos observar la correlacion es negativa los valores altos de una de ellas le suelen corresponder valor bajos de la otra y viceversa, ambas variables tienden a disminuir a la vez estas variables no son independientes ya que la correlacion no es igual acero.
