```
data = pd.read_csv('gdrive/My Drive/parcialcte1/fc2-exam-primercorte2020i-ogils/Casos1_13abr2020.csv') # Leyendo datos desde el archivo Casos1.csv
```

Numero de casos covid-19 por Sexo y Edad

```
dataSex = pd.DataFrame(data, columns= ['Fecha de diagnóstico','Sexo','Edad']) # Cargando en la variable dataSex los campos: Fecha diagnóstico, Sexo y Edad
```

```
histo = dataSex.groupby('Edad')['Sexo'].sum() # Suma el número de apariciones de M y F: Masculino, Femenino

print(histo.str.count('F'), histo.str.count('M'))
```

```
male = histo.str.count('M')
female = histo.str.count('F')
total = male.sum()+female.sum()

print("hombres:",male.sum(), round((male.sum()/total)*100,2), "%")
print("Mujeres:",female.sum(), round((female.sum()/total)*100,2), "%")
print("Total:", total)
```

```
plot(histo.str.count('M'),'bx')
plot(histo.str.count('F'),'r.')
xticks(rotation=45)
grid()
show()
```

```
datatip = pd.DataFrame(data, columns= ['Fecha de diagnóstico','Tipo*','Edad'])
histo = datatip.groupby('Edad')['Tipo*'].sum()
```

Numero de casos por Tipo y Edad

```
print(histo.str.count('Relacionado'), histo.str.count('Importado'))
```

```
Rela = histo.str.count('Relacionado')
Impor = histo.str.count('Importado')
total= Rela.sum()+Impor.sum()
print("Tipo relacionado:", Rela.sum(), round(( Rela.sum()/total)*100,2), "%")
print("Tipo importado:",Impor.sum(), round((Impor.sum()/total)*100,2), "%")
print("Total:", total)
```

```
plot(histo.str.count('Relacionado'),'gx')
plot(histo.str.count('Importado'),'r.')
xticks(rotation=45)
grid()
show()
```
